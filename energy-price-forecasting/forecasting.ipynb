# Energy Price Forecasting
This notebook forecasts energy costs and consumption using ARIMA and Prophet.

```
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Load cleaned dataset
df = pd.read_csv('../energy-etl-pipeline/data/sample_clean.csv')
df = df[['period', 'cost']].rename(columns={'period':'ds', 'cost':'y'})
df['ds'] = pd.to_datetime(df['ds'])

# Train Prophet model
model = Prophet()
model.fit(df)

# Forecast future
future = model.make_future_dataframe(periods=12, freq='M')
forecast = model.predict(future)

# Plot forecast
fig = model.plot(forecast)
plt.savefig('visuals/forecast_plot.png')
plt.show()
```
